```python
import pandas as pd

# pd.read_csv('파일 경로')
Company = pd.read_csv('/content/Company.csv')
```

```python
Company.head()
```

```python
# Item1, Item2, Item3을 모두 더한
# 새로운 칼럼 Item_sum 만들기
Company["Item_sum"] = Company["Item1"] + Company["Item2"] + Company["Item3"]
```

```python
Company.head()
```

```python
# Item1, Item2, Item3의 평균을
# 새로운 칼럼 Item_mean에 저장
# = Item_sum / 3
Company["Item_mean"] = Company["Item_sum"] / 3
```

```python
Company
```

# 조건에 따라 다른 값을 갖는 새로운 칼럼 만들기

```python
# 조건을 만족하는 개체 선택하기
# df.loc[조건, 칼럼들]
# 조건 1: AgeGroup이 3 이하이면 "30대 이하"
# 조건 2: AgeGroup이 4 이상이면 "40대 이상"
# 두 조건에 따라 값을 새로운 칼럼 Type에 저장
cond1 = (Company["AgeGroup"] <= 3)
cond2 = (Company["AgeGroup"] >= 4)
```

```python
Company.loc[cond1, "Type"] = "30대 이하"
```

```python
Company
```

```python
# AgeGroup이 4이상인 경우 "Type"에 "40대 이상"
Company.loc[cond2, "Type"] = "40대 이상"
```

```python
Company
```

# Gender 값이 F면 여성, M이면 남성이라는 값을 가지는 새로운 변수 생성하기

```python
# 방법 1
# cond3 = (Company["Gender"] == "F")
# cond4 = (Company["Gender"] == "M")
# Company.loc[cond3, "Gender_kor"] = "여성"
# Company.loc[cond4, "Gender_kor"] = "남성"
```

```python
cond3 = (Company["Gender"] == "F")
cond4 = (Company["Gender"] == "M")
Company.loc[cond3, "Gender_kor"] = "여성"
Company.loc[cond4, "Gender_kor"] = "남성"
```

# df.groupby("뭘로 그룹을 만들 것인지, 칼럼 명").agg("집계함수")

```python
# Gender로 그룹 만들기
Company.groupby("Gender")
```

```python
# Gender로 그룹 만들기
# "ID" 개수 세보기
# df.groupby().agg({"칼럼":"집계함수"})
Company.groupby("Gender").agg({"ID":"count"})
```

```python
# Gender로 그룹 만들기
# "ID" 개수 세보기
# "Item1"의 평균
# df.groupby().agg({"칼럼":"집계함수"})
Company.groupby("Gender").agg({"ID":"count",
                               "Item1":"mean"})
```

```python
Company.groupby("Gender").agg({"ID":"count",
                               "Item1":"mean",
                               "Item3":"min"})
```
