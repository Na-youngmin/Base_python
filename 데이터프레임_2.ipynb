## 대부분의 프로그래밍 언어에서 "같다"는 의미는 등호 2개로 사용 "=="

```python
# Company.loc[행 선택 값, 열 선택 값]
# AgeGroup이 4인 행들
# 전체 열
Company.loc[Company["AgeGroup"] == 4, :]
```

```python
Company["AgeGroup"] >= 2 # 2이상
```

```python
# AgeGroup이 2이상인 행들
# 같다 ==
# 이상 >=
# 이하 <=
# 전체 열
Company.loc[Company["AgeGroup"] >= 2, :]
```

```python
# Gender가 M인 행들
# 전체 열
Company["Gender"] == "M"
```

```python
Company.loc[Company["Gender"] == "M", :]
```

```python
# 같다 ==
# 같지 않다 !=
# 이상 >=
# 이하 <=
# Gender가 M이 아닌 행들
# 전체 열
Company.loc[Company["Gender"] != "M", :]
```

```python
# AgeGroup이 4이거나 5인 개체들 선택하기
# 조건 1: Company["AgeGroup"] == 4
# 조건 2: Company["AgeGroup"] == 5
# 두 개의 조건을 묶는 법
# 그리고(and) &
# 또는(or) | <- vertical bar, 엔터키 옆에 있는 애
```

```python
(Company["AgeGroup"] == 4) | (Company["AgeGroup"] == 5)
```

```python
Company.loc[(Company["AgeGroup"] == 4) | (Company["AgeGroup"] == 5),
            :]
```

```python
# AgeGroup이 2이상 4이하 개체들 선택하기
# 조건 1: Company["AgeGroup"] >= 2
# 조건 2: Company["AgeGroup"] <= 4
# 두 개의 조건을 묶는 법
# 그리고(and) &
# 또는(or) | <- vertical bar, 엔터키 옆에 있는 애
```

```python
(Company["AgeGroup"] >= 2) & (Company["AgeGroup"] <= 4)
```

```python
cond1 = Company["AgeGroup"] >= 2
cond2 = Company["AgeGroup"] <= 4
cond = cond1 & cond2
```

```python
Company.loc[cond, :]
```
